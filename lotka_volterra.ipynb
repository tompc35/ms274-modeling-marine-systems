# Lotka-Volterra model

Classic predator-prey model. Default parameter values for phytoplankton-grazer interaction are from C. B. Miller's *Biological Oceanography* text.


$dP/dt = bP - aPZ$

$dZ/dt =  -dZ + kaPZ$


```
import numpy as np
import matplotlib.pyplot as plt
```

```
# parameters
b = 1
a = 0.1
d = 0.75
k = 0.1

# initial conditions
P0 = 75
Z0 = 25

# set up time steps
tmax = 120
dt = 0.01

# initialize arrays
t = np.arange(0,tmax,dt)
nsteps = len(t)
P = np.zeros(nsteps)
Z = np.zeros(nsteps)
P[0] = P0
Z[0] = Z0

# loop through to solve equations with Euler's method
for i in range(0,nsteps-1):
  dPdt = b*P[i] - a*P[i]*Z[i]
  dZdt = -d*Z[i] + k*a*P[i]*Z[i]
  P[i+1] = P[i] + dPdt*dt
  Z[i+1] = Z[i] + dZdt*dt
```

### Plot time series

```
plt.figure()
plt.plot(t,P)
plt.plot(t,Z)
plt.xlabel('time')
plt.ylabel('abundance')
plt.legend(['P','Z'])
```

### Phase plane diagram

```
plt.figure()
plt.scatter(P,Z,6,t,cmap='viridis')
plt.colorbar(label='time')
plt.xlabel('P')
plt.ylabel('Z')
```
